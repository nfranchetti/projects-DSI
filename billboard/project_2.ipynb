### In the next few cells, I rename the columns corresponding to the weekly ranking of the songs. For example, the 'x1st.week' column becomes 'week1'. This will help with sorting the rankings in chronological order later on.

```python
week_columns = []
counter = 1
for column in df.columns:
    if column[0] == 'x':
        week_columns.append('week'+ str(counter))
        counter += 1
    else:
        week_columns.append(column)
        
week_columns


df.columns = week_columns
```

```python
df.columns
```

### Below, I convert the length of the song from minutes:seconds to seconds (ie: 3:38 to 218 seconds). This will allow me to make comparisons and create visualizations using the track length.

```python
print df['time'][0]
```

```python
def time_splitter(time):
    new_time = time.split(':')
    return int(new_time[0])*60 + int(new_time[1])

for i in range(len(df['time'])):
#     print df['time'][i]
    df['time'][i] = time_splitter(df['time'][i])
#     print df['time']
#     print i
```

```python
df['time']
```

```python
# Converting the date.entered and date.peaked to pandas dataframe datetime type
df['date.entered'] = pd.to_datetime(df['date.entered'],infer_datetime_format=True)
df['date.peaked'] = pd.to_datetime(df['date.peaked'],infer_datetime_format=True)
```

```python
df.dtypes
```

```python
df.head()
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
# Separating the colums into info columns, which list info about the song, and week columns, which give the ranking for each week.
week_cols = []
info_cols = []
for col in df:
    if col[0] == 'w':
        week_cols.append(col)
    else:
        info_cols.append(col)

info_cols
```

```python
# Using pd.melt to created a melted dataframe, with each weekly ranking for each song having its own row.
melted_df = pd.melt(df, id_vars=info_cols, value_vars=week_cols, var_name='Week', value_name='Ranking')
melted_df
```

```python
# The dataframe above still contains all the NaN rows from weeks that don't have data for a specific song.
# Here I drop all the NaN rows
no_nan_melted_df = melted_df.dropna()
no_nan_melted_df
```

```python
# This is just a sanity check to see if the amount of data for each week makes sense and match the value counts I did above
no_nan_melted_df['Week'].value_counts()
```

```python
# This creates a pivot table that will show the info for each song, with a row for each week and the corresponding ranking
pivot_melted_df = pd.pivot_table(no_nan_melted_df, values=['Ranking'], index=['artist.inverted', 'track', 'time', 'genre', 'date.entered', 'date.peaked', 'Week'])#.reset_index()
pivot_melted_df.head(50)
```
